```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
df = pd.read_csv('dataset_olympics.csv')
```

```python
df
```

```python
df = df.drop_duplicates()
```

```python
df
```

```python
olympics_data = df.drop(['Height','Weight'],axis = 1)
```

```python
olympics_data
```

```python
olympics_data.isnull().sum()
```

```python
olympics_data['Medal'] = olympics_data['Medal'].fillna('Medal-less')
```

```python
olympics_data
```

```python
olympics_data['Age'] = olympics_data['Age'].fillna(olympics_data['Age'].median())
```

```python
olympics_data.isnull().sum()
```

```python
olympics_data
```

```python
olympics_data.to_csv('olympics_history_cleaned_data.csv', index=False)
```
